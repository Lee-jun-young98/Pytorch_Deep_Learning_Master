```python
import torch
import cv2
blue = cv2.imread("blue.jpg")
red = cv2.imread("red.jpg")
green = cv2.imread("green.jpg")
```

```python
blue = torch.Tensor(blue)
red = torch.Tensor(red)
green = torch.Tensor(green)
```

```python
blue = blue.permute(2,0,1)
red = red.permute(2,0,1)
green = green.permute(2,0,1)
```

```python
## BGR 순서로 저장되어 있음

print(blue.mean())
print(red.mean())
print(green.mean())
```

```python
blue.mean(dim=2).mean(dim=1)
```

```python
red.mean(dim=2).mean(dim=1)
```

```python
green.mean(dim=2).mean(dim=1)
```
